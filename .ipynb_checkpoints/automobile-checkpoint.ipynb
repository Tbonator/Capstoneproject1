```python
import numpy as np
import pandas as pd
import seaborn as sns
import ast, json

from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline


automobile_df = pd.read_csv("automobile.txt")

```

Check the head of the DataFrame.

```python
automobile_df.head()
```

```python
#How many rows and columns are there?
automobile_df.columns
```

```python
#  dropping the duplicates from the dataset.
print(automobile_df.shape)
automobile_df = automobile_df.drop_duplicates(keep='first')
print(automobile_df.shape)
```

```python
automobile_df.describe()
```

```python
# checking missing data
automobile_df.isnull().any().any()
```

```python
#data information 
automobile_df.info()
```

```python
# replacing ? with NAN
automobile_df = automobile_df.replace('?',np.NAN) 
automobile_df.isnull().sum()
```

```python
df_temp = automobile_df[automobile_df['normalized-losses']!='?']
normalised_mean = df_temp['normalized-losses'].astype(float).mean()
automobile_df['normalized-losses'] = automobile_df['normalized-losses'].replace('?',normalised_mean).astype(float)

df_temp = automobile_df[automobile_df['price']!='?']
normalised_mean = df_temp['price'].astype(float).mean()
automobile_df['price'] = automobile_df['price'].replace('?',normalised_mean).astype(float)

df_temp = automobile_df[automobile_df['horsepower']!='?']
normalised_mean = df_temp['horsepower'].astype(float).mean()
automobile_df['horsepower'] = automobile_df['horsepower'].replace('?',normalised_mean).astype(float)

df_temp = automobile_df[automobile_df['peak-rpm']!='?']
normalised_mean = df_temp['peak-rpm'].astype(float).mean()
automobile_df['peak-rpm'] = automobile_df['peak-rpm'].replace('?',normalised_mean).astype(float)
```

```python
automobile_df.head()
```

```python
automobile_df[['engine-size','peak-rpm','curb-weight','horsepower','price']].hist(figsize=(10,8),bins=6,color='g')
# 2 plt.figure(figsize=(10,8))
plt.tight_layout()
plt.show()
```

```python
plt.figure(1)
plt.subplot(221)
automobile_df['engine-type'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='red')
plt.title("Number of Engine Type frequency diagram")
plt.ylabel('Number of Engine Type')
plt.xlabel('engine-type');


plt.subplot(222)
automobile_df['num-of-doors'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='green')
plt.title("Number of Door frequency diagram")
plt.ylabel('Number of Doors')
plt.xlabel('num-of-doors');

plt.subplot(223)
automobile_df['fuel-type'].value_counts(normalize= True).plot(figsize=(10,8),kind='bar',color='purple')
plt.title("Number of Fuel Type frequency diagram")
plt.ylabel('Number of vehicles')
plt.xlabel('fuel-type');

plt.subplot(224)
automobile_df['body-style'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='orange')
plt.title("Number of Body Style frequency diagram")
plt.ylabel('Number of vehicles')
plt.xlabel('body-style');
plt.tight_layout()
plt.show()
```
